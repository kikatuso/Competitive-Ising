```python

import networkx as nx 
import math  
import torch 
from torch.autograd import grad as torchgrad

from src import helperfunctions as hf

class mf_ising_system:

    def __init__(self,graph,background_field,fixed_point_iter=1000,init_mag='random',fp_tol_fac=10-5,
        iim_iter=1000,iim_tol_fac=1e-5,momentum=0.4,optimiser_type='sgdm'):
        
        self.graph = graph
        self.adj_matrix = torch.from_numpy(nx.to_numpy_matrix(graph))
        self.graph_size = len(graph.nodes.keys())
        self.background_field = background_field
        self.fixed_point_iter=fixed_point_iter
        self.iim_iter = iim_iter
        self.fp_tol_fac=fp_tol_fac
        self.iim_tol_fac=iim_tol_fac
        self.optimiser_type = optimiser_type
        self.momentum = momentum
        if init_mag=='aligned_neg':
            self.init_mag=(-1)*torch.ones(self.graph_size,dtype=torch.float64)
        if init_mag=='aligned_pos':
            self.init_mag=torch.ones(self.graph_size,dtype=torch.float64)
        if init_mag=='random':
            self.init_mag=torch.tensor([np.random.choice([-1,1]) for i in range(self.graph_size)],dtype=torch.float64)
        else:
            self.init_mag = init_mag 


            
        
    def single_mag(self,i,m,beta,field):
        gamma=1.0
        spin_field = torch.dot(self.adj_matrix[i],m)
        term = torch.tanh(beta*(spin_field+field[i]))
        return (1.0-gamma)*m[i] + gamma*term
    
    def magnetisation(self,mag,beta,field):
        m_old = mag
        m_new = torch.zeros(len(m_old),dtype=torch.float64)
        for i in range(self.graph_size):
            m_new[i]=self.single_mag(i,m_old,beta,field)
        return m_new


    
    def positive_agent(self,mag_i,it,pos_budget,beta,jt):
        mag_i_grad = self.mag_grad(beta,mag_i)
        
        control_field_update = control_field + self.step_size*mag_i_grad
 
        control_field_new = hf.projection_simplex_sort(control_field_update.T,z=pos_budget)
        self.control_field_history_pos.append(control_field_new)
        
        last_control_field_neg = self.control_field_history_neg[-1] # adding contribution from the negative agent
        tot_field = np.array(self.background_field+last_control_field_neg+control_field_new)
        mag_ii= self.Steffensen_method(mag_i,beta,tot_field,jt)
        return mag_ii
    
    
    def negative_agent(self,mag_i,it,neg_budget,beta,jt):
        mag_i_grad = self.mag_grad(beta,mag_i)
        control_field = self.control_field_history_neg[it]

        if self.optimiser_type=='sgd':
            control_field_update = control_field - self.step_size*mag_i_grad

            
        control_field_new = -hf.projection_simplex_sort(control_field_update.T,z=neg_budget)
        self.control_field_history_neg.append(control_field_new)
        
        last_control_field_pos = self.control_field_history_pos[-1]        
        tot_field = np.array(self.background_field+control_field_new+last_control_field_pos)
        mag_iii= self.Steffensen_method(mag_i,beta,tot_field,jt)
        return mag_iii
    
    def nth_derivative(self,f, wrt, n):
        for i in range(n):
            grads = -torchgrad(f, wrt, create_graph=True)[0]
            f = grads.sum()
        return grads

        
    def MF_IIM(self,H_pos,H_neg,beta,budget_pos,budget_neg,lr_fast,lr_slow):
               

        self.control_field_history_pos = []
        self.control_field_history_neg = []
        self.mag_delta_history = []
        
        control_pos = H_pos.data
        control_neg = H_neg.data
        self.control_field_history_pos.append(control_pos.detach().numpy())
        self.control_field_history_neg.append(control_neg.detach().numpy())
        self.mag_delta_history.append(self.init_mag.detach().numpy())
        
        self.changes = [np.zeros(self.graph_size)]
        previous_mag = self.init_mag
        for it in range(self.iim_iter):
            total_field=H_pos + H_neg 
            
            mag = self.magnetisation(previous_mag,beta,total_field)

            self.mag_delta_history.append(mag.detach().numpy())
            
            mag.backward(torch.ones_like(H_pos),retain_graph=True,create_graph=True)
            
            D1f1 = H_pos*H_pos.grad
            D2f2 = -H_neg*H_neg.grad
            D2f1 = -D2f2
            D21f2 = torch.tensor([float(self.nth_derivative(D2f2[i],H_pos,1)[i].detach()) for i in range(self.graph_size)])
            D22f2 = torch.tensor([float(self.nth_derivative(D2f2[i],H_neg,1)[i].detach()) for i in range(self.graph_size)])
            
            H_pos_update = D1f1 - D2f1*(D21f2/D22f2)
            H_neg_update = D2f2

            
            H_pos.data.add_(lr_fast* H_pos_update)
            H_neg.data.add_(lr_slow * H_neg_update)
            H_pos.data = hf.projection_simplex_sort_torch(H_pos, z=budget_pos)
            H_neg.data = -hf.projection_simplex_sort_torch(H_neg, z=budget_neg)
            H_pos.grad.detach_()
            H_neg.grad.detach_()


            
            h_pos = H_pos.detach().numpy()
            h_neg = H_neg.detach().numpy()

            self.control_field_history_pos.append(h_pos)
            self.control_field_history_neg.append(h_neg)

            previous_mag = mag
               
            if torch.abs(H_pos.grad.data.sum())<self.iim_tol_fac and D22f2.sum()>0:
                print(it)
                break
            
            H_pos.grad.zero_()
            H_neg.grad.zero_()
        if it==self.iim_iter-1:
            print('Failed to converge after {} iterations'.format(self.iim_iter))
            final_mag = None
            
        elif it < self.iim_iter-1:
            final_mag = mag.detach().numpy()
        
        self.control_field_history_pos = np.array(self.control_field_history_pos)
        self.control_field_history_neg = np.array(self.control_field_history_neg)
        return H_pos.detach().numpy(),H_neg.detach().numpy(),final_mag
```
